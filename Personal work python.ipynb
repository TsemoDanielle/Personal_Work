## FUSION DE PLUSIEURS PDF EN UN SEUL DATAFRAME

```python

import pandas as pd
import PyPDF2 as pypdf
def fusion():
    termine="oui"
    data= pd.DataFrame()
    while termine!="non":
        
        chemin=input("Entrez le chemin d'accès vers le pdf: ")
        pdftest=open(chemin,'rb')
        pdf=pypdf.PdfReader(pdftest)
        test = pdf.get_form_text_fields()
        dfs = pd.DataFrame.from_dict(test, orient = "index")
        dd = dfs.transpose()
        col = list(dd.columns)
        col= col[-1::-1]
        dd = dd[col]
        data= pd.concat([data, dd], axis=0)
        termine=input("Voulez vous ajouter un autre pdf ? oui/non ")
    
    return data
        
```

```python
fusion()
```
